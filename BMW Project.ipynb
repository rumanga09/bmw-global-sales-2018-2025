```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Project-PS 23-Muhammad Arya Rumanga/bmw_global_sales_2018_2025.csv")

df.head()
df.info()
```

```python
cols = ['Units_Sold', 'Avg_Price_EUR', 'Revenue_EUR', 'BEV_Share']
summary = df[cols].agg(['mean', 'median', 'std']).T
summary.columns = ['Mean', 'Median', 'Std Dev']
print(summary)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df["Date"] = pd.to_datetime(
    dict(
        year=df["Year"],
        month=df["Month"],
        day=1
    )
)
```

```python
df.head()
```

```python
df["Vehicle_Type"] = df["Model"].apply(
    lambda x: "EV" if x in ["i4","iX"] else "ICE"
)
```

```python
df.to_csv(
    "bmw_sales_clean.csv",
    index=False
)
```

```python
# korelasi langsung
r1 = df['GDP_Growth'].corr(df['BEV_Share'])
r2 = df['Fuel_Price_Index'].corr(df['BEV_Share'])
r4 = df['GDP_Growth'].corr(df['Avg_Price_EUR'])

# Filter hanya ICE
ice = df[df['Vehicle_Type'] == 'ICE']
r3 = ice['Fuel_Price_Index'].corr(ice['Units_Sold'])

print(f"GDP vs BEV Share: {r1:.3f}")
print(f"Fuel Price vs BEV Share: {r2:.3f}")
print(f"Fuel Price vs ICE Units: {r3:.3f}")
print(f"GDP vs Avg Price: {r4:.3f}")
```
